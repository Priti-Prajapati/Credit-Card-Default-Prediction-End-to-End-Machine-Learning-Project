```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\PRITI\.cache\kagglehub\datasets\uciml\default-of-credit-card-clients-dataset\versions\1\UCI_Credit_Card.csv')
```

```python
sns.countplot(x='default.payment.next.month', data=df)
plt.title("Target Variable Distribution")
plt.show()
```

```python
education_level={
    1: "Graduate School",
    2: "University",
    3: "High School",
    4: "Others",
    5: "Unknown",
    6: "Unknown"}
```

```python
df['EDUCATION_LABEL'] = df['EDUCATION'].map(education_level)
```

```python
sns.countplot(x='EDUCATION_LABEL', data=df)
plt.title("Education Level")
plt.xticks(rotation=45)
plt.show()
```

```python
marriage_status = {
    1: "Married",
    2: "Single",
    3: "Others"
}
```

```python
df['MARRIAGE_LABEL'] = df['MARRIAGE'].map(marriage_status)
```

```python
marriage_counts = df['MARRIAGE_LABEL'].value_counts()
```

```python
plt.figure(figsize=(3,3))
plt.pie(marriage_counts, labels=marriage_counts.index, autopct='%1.2f%%')
plt.title("Marital Status Distribution")
plt.show()
```

```python
gender_count = {
    1: "Male",
    2: "Female"
}

df['GENDER_LABEL'] = df['SEX'].map(gender_count)
```

```python
sns.countplot(y='GENDER_LABEL', data=df,palette={'Male': 'skyblue', 'Female': 'pink'})
plt.title("Gender Distribution")
plt.xlabel("Count")
plt.ylabel("Gender")
plt.show()
```
